```python
despicable_bgr = cv2.imread('despicable.webp', cv2.IMREAD_COLOR)
```

```python
b, g, r = cv2.split(despicable_bgr)
```

```python
plt.figure(figsize=(20,8))
plt.subplot(141); plt.imshow(r, cmap='gray'); plt.title('Red Channel');
plt.subplot(142); plt.imshow(g, cmap='gray'); plt.title('Green Channel');
plt.subplot(143); plt.imshow(b, cmap='gray'); plt.title('Blue Channel');

despicable_merged = cv2.merge((b, g, r))

plt.subplot(144); plt.imshow(despicable_merged[:,:,::-1]); plt.title('Merged Output')
```

```python
superbowl = cv2.imread('superbowl.webp', 1)
superbowl = superbowl[:,:,::-1]
```

```python
plt.imshow(superbowl)
plt.title('Matplotlib IM show')
plt.show()

window1 = cv2.namedWindow('W1')
cv2.imshow(window1, superbowl)
cv2.waitKey(3000)
cv2.destroyWindow(window1)

window2 = cv2.namedWindow('W2')
cv2.imshow(window2, superbowl)
cv2.waitKey(3000)
cv2.destroyWindow(window2)

window3 = cv2.namedWindow('W3')
cv2.imshow(window3, superbowl)
cv2.waitKey(0)
cv2.destroyWindow(window3)
```
